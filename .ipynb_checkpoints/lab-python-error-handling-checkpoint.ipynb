1. Define the function for initializing the inventory with error handling

````python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory
    
inventory = initialize_inventory(products)
print(products)
````

2. Modify the calculate_total_price function to include error handling.

If the user enters an invalid price (e.g., a negative value or a non-numeric value), display an error message and ask them to re-enter the price for that product.
Use a try-except block to handle the error and continue prompting the user until a valid price is entered.


````python
total_products_ordered={'t-shirt': 3, 'mug': 1}
````

````python
def calculate_order_price(inventory,total_products_ordered):
    inventory_cost = {}
    
    for product in inventory:
        isValid = False
        while not isValid:
            try:
                price = int(input(f"Unit price of {product}: "))
                if price < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                isValid = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory_cost[product]= price

    cost_of_order = sum(inventory_cost[product] * quantity for product,quantity in total_products_ordered.items())
        
    return cost_of_order,inventory_cost

cost_of_order, inventory_cost  = calculate_order_price(inventory,total_products_ordered)

print(cost_of_order)
print(inventory_cost)
````

3. Modify the get_customer_orders function to include error handling.

If the user enters an invalid number of orders (e.g., a negative value or a non-numeric value), display an error message and ask them to re-enter the number of orders.
If the user enters an invalid product name (e.g., a product name that is not in the inventory), or that doesn't have stock available, display an error message and ask them to re-enter the product name. Hint: you will need to pass inventory as a parameter
Use a try-except block to handle the error and continue prompting the user until a valid product name is entered.


````python
def get_customer_orders():
    total_products_ordered = {}

    #Ask for number of different items to order
    isValid = False
    while not isValid:
        try:
            num_orders = int(input(f"How many different products would you like to order? "))
            if num_orders < 0:
                raise ValueError("Invalid quantity! Please enter a non-negative value.")
            isValid = True
        except ValueError as error:
            print(f"Error: {error}")

    #Get the number of items per desired product        
    for i in range(num_orders):
        isValid = False
        while not isValid:
            try:
                product = input(f"Enter product #{i + 1}: ").strip().lower()
                if product not in products:
                    raise ValueError(f"Invalid product. Choose from: {', '.join(products)}")
                ordered_quantity = int(input("Quantity: "))
                if ordered_quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                total_products_ordered[product] = ordered_quantity
                isValid = True
            except ValueError as error:
                print(f"Error: {error}")
      
    return total_products_ordered
    
#num_orders= get_customer_orders()
total_products_ordered = get_customer_orders()
````
